```python
import numpy as np
import matplotlib.pyplot as plt
import astropy.units as u
from astropy.time import Time
import healpy as hp
import zodipy
%matplotlib inline
```

```python
model = zodipy.Zodipy()
```

```python
def reset_emissivities(model):
    parameteres = model.get_parameters()
    comps = parameteres['emissivities'].keys()
    for comp in comps:
        parameteres['emissivities'][comp] = np.ones_like(parameteres['emissivities'][comp])
    model.update_parameters(parameteres)
reset_emissivities(model)
```

```python
nside = 32

emission = model.get_binned_emission_pix(
    freq=25*u.micron,
    pixels=np.arange(hp.nside2npix(nside)),
    nside=nside,
    obs_time=Time.now(),
)
```

```python
hp.mollview(emission, min=0, max=80)
```
